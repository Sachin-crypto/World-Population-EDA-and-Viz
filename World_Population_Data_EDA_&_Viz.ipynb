# **EDA**

```python
# Most populated country
most_populated = df[df['2022 Population'] == max(df['2022 Population'])]
most_populated
```

```python
# Least populated country
least_populated = df[df['2022 Population'] == min(df['2022 Population'])]
least_populated
```

```python
# Densely populated country
density = df[df['Density (per km²)'] == max(df['Density (per km²)'])]
density
```

```python
# Top 5 densely populated countries
dense = df.nlargest(n=5, columns='Density (per km²)')
fig = px.bar(dense, x="Country", 
             y=['Density (per km²)'], 
             title="Top 5 densely populated countries",
             labels={"value": "Density (per km²)", "variable": "Density"})
fig.show()
```

```python
# Top 5 least densely populated countries
dense = df.nsmallest(n=5, columns='Density (per km²)')
fig = px.bar(dense, x="Country", 
             y=['Density (per km²)'], 
             title="Top 5 least densely populated countries",
             labels={"value": "Density (per km²)", "variable": "Density"})
fig.show()
```

```python
# Top 5 highly populated countries
pop = df.nlargest(n=5, columns='2022 Population')
fig = px.bar(pop, x="Country", 
             y=['1970 Population', '2022 Population'], 
             title="Top 5 highly populated countries",
             labels={"value": "Population", "variable": "Population"})
fig.show()
```

```python
# Top 5 least populated countries
pop = df.nsmallest(n=5, columns='2022 Population')
fig = px.bar(pop, x="Country", 
             y=['1970 Population', '2022 Population'], 
             title="Top 5 least populated countries",
             labels={"value": "Population", "variable": "Population"})
fig.show()
```

```python
# Top 15 countries having largest area
area = df.nlargest(n=15, columns='Area (km²)')

fig = px.bar(area, x="Country", 
             y=['Area (km²)'], 
             title="Countries having largest area",
             labels={"value": "Area (km²)", "variable": "Area (km²)"})
fig.show()
```

```python
# Top 15 countries having smallest area
area = df.nsmallest(n=15, columns='Area (km²)')

fig = px.bar(area, x="Country", 
             y=['Area (km²)'], 
             title="Countries having smallest area",
             labels={"value": "Area (km²)", "variable": "Area (km²)"})
fig.show()
```
